```python
import os
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt 
from matplotlib import dates
%matplotlib inline
```

```python
users_df = pd.read_csv('data/yammer_users.csv')
users_df.head(10)
```

```python
events_df = pd.read_csv('data/yammer_events.csv')
# Load and convert date time into weeks
events_df['occurred_at_period'] = pd.to_datetime(events_df['occurred_at'])
events_df['occurred_at_period'] = events_df['occurred_at_period'].dt.to_period(freq = 'W')

events_df['occurred_at_period_start_time'] = events_df.occurred_at_period.dt.start_time
events_df['occurred_at_period_end_time'] = events_df.occurred_at_period.dt.end_time

# drop user_type due to too many nans
events_df.isna().sum()
events_df.drop(['occurred_at_period', 'user_type'], axis=1, inplace=True)

events_df.head(10)
```

```python
roll_up_period = pd.read_csv("data/dimension_rollup_periods.csv")
roll_up_period['occurred_at_period_start_time'] = pd.to_datetime(roll_up_period['pst_start'])
roll_up_period.head()
```

### 1. Merging dataframes [users + events]

```python
# lets first join all the dataframes.
events_users_df = pd.merge(events_df, users_df, on = "user_id", how="left")
events_users_df.head()
```

```python
# Keep only engaged for plotting
engaged = events_users_df[events_users_df.event_type == "engagement"]
engaged_group = engaged.groupby('occurred_at_period_start_time')['user_id'].nunique().to_frame().reset_index()
#engaged_group.head()

# Plotting
fig = plt.subplots(1,1, figsize = (13, 5))
ax = sns.lineplot(data=engaged_group, x="occurred_at_period_start_time", y="user_id")
ax.set(xticks=engaged_group.occurred_at_period_start_time.values);
ax.xaxis.set_major_formatter(dates.DateFormatter("%d-%b-%Y"));
plt.xticks(rotation=90);
plt.xlabel("");
plt.ylabel("");
plt.title("User engagment from April 2014 to August 2014");
```

```python
email_df = pd.read_csv('data/yammer_emails.csv')
email_df.head(10)
```

```python
from datetime import datetime

#Convert column data types
from datetime import datetime
email_df['occurred_at']= email_df['occurred_at'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d %H:%M:%S'))
email_df['action']= email_df['action'].astype("category")
email_df['user_type']=email_df['user_type'].astype("category")
email_df['user_id']=email_df['user_id'].astype("category")

print(len(set(email_df['action'])))
print(len(set(email_df['user_id'])))
print(len(set(email_df['occurred_at'])))
print(len(set(email_df['user_type'])))
print(email_df.head())
```

```python
email_df = email_df.set_index('occurred_at')
grouped = email_df.groupby('user_type')

fig = plt.subplots(1,1, figsize = (13, 5))
for key, group in grouped:
    data = group.groupby(lambda x: x.date).count()
    data['user_type'].plot(label=key)

#plt.xlim(2014, 2014)
#plt.ylim(0,5000)
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
grouped = email_df.groupby('action')

fig = plt.subplots(1,1, figsize = (13, 5))
for key, group in grouped:
    data = group.groupby(lambda x: x.date).count()
    data['action'].plot(label=key)

#plt.ylim(0,5000)
plt.xticks(rotation=90)
plt.legend()
plt.show()

fig = plt.subplots(1,1, figsize = (13, 5))
for key, group in grouped:
    data2 = group.groupby(lambda x: x.month).count()
    data2['action'].plot(label=key)

#plt.ylim(0,5000)
plt.legend()
plt.show()
```
